```julia
using DifferentialEquations
using Plots, LaTeXStrings
using DataInterpolations
```

```julia
d = collect(0:0.05:1)
data = sin.(5*2pi.*d)
intp = LinearInterpolation(data,d);
```

```julia
plot(d,intp.(d))
plot!(d,data)
```
